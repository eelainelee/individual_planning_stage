**<h4>(2) Questions</h4>**

<u>Broad question</u>: What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?


<u>Specific question</u>: Can the age and playtime of players predict who will subscribe to game-related newsletter?


The `played_hours`, `Age`, and `subscribe` variables from the `players.csv` dataset will be used to answer this question.

**<h4>(3) Exploratory Data Analysis and Visualization</h4>**

<h5><u>players.csv data analysis</u></h5>

```r
# loading data into R
players <- read_csv("https://raw.githubusercontent.com/eelainelee/individual_planning_stage/refs/heads/main/players.csv")

# no need to tidy data

# computing mean values
mean_played_hours <- round(summarize(players, mean(played_hours)),2)
mean_played_hours

mean_Age <- round(summarize(players, mean(Age, na.rm = TRUE)),2)
mean_Age

# exploratory visualization
options(repr.plot.width = 7, repr.plot.height = 7)
players_plot <- ggplot(players, aes(x = Age, y = played_hours, color = subscribe)) +
    geom_point() + 
    labs(x = "Age (years)",
         y = "Played time (hours)",
         color = "Subscribed",
         title = "Relation of player age and playtime to newsletter subscription")
    
players_plot 
```

**<h5>Insights gained from visualization</h5>**


Upon generating a scatterplot of playtime vs. age, there appears to be no distinct relationship between the two variables. However, this graph suggests that the most common age group of players ranges from approximately 17-26 years old. The playtime of most players appears very low, yet precise values cannot be determined from this visualization. By identifying and omitting major outliers to reduce the size of both axes, a new plot can be generated to view all points with greater accuracy. Although the relationship between the 2 quantitative values is difficult to view from this plot, it can be observed that players with higher playtime hours are all subscribed to the game-related newsletter. This current relationship, which is clearly visible, is subject to change once outliers are omitted; hence, the significance of this observation remains unknown until a new visualization is generated.
